# 02. Chat Model 호출

LangChain에서 OpenAI ChatModel을 불러오고 기본 호출 방법을 실습합니다.

## 1. 환경 설정

```python
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path="../.env")
```

## 2. ChatOpenAI 초기화

- `model`: 사용할 모델 이름
- `temperature`: 0에 가까울수록 일관된 답변, 1에 가까울수록 창의적인 답변

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
print(llm)
```

## 3. 기본 호출 - invoke()

문자열을 직접 전달하면 HumanMessage로 자동 변환됩니다.

```python
question = "현재 대한민국 대통령 누구야?"
response = llm.invoke(question)
print(response)
```

```python
# content만 출력
#컴퓨터 메모리에 저장되어있기 때문에 다시 요청하지않음

print(response.content)
```

## 4. 메시지 객체로 호출

역할(role)을 구분하여 메시지를 전달할 수 있습니다.
- `SystemMessage`: AI의 역할/페르소나 설정
- `HumanMessage`: 사용자 입력

```python
from langchain_core.messages import SystemMessage, HumanMessage

messages = [
    SystemMessage(content="너는 행정학 교수야."),
    HumanMessage(content="현재 대한민국 대통령 누구야?")
]


response = llm.invoke(messages)
print(response.content)
```

- 마크다운으로 출력해서 보기

```python
from IPython.display import Markdown, display
display(Markdown(response.content))
```

## 5. 스트리밍 호출 - stream()

- 토큰이 생성되는 즉시 출력함.
- llm 객체에 질문을 보내고, 응답을 한 번에 받지 않고 조각(chunk) 단위로 순차적으로 받음.
- end=""를 쓰면 줄바꿈 없이 이어서 출력함.
- flush=True는 출력 버퍼를 즉시 비우라는 의미임.

```python
for chunk in llm.stream("최신 langchain 버전을 알려줘 "):
    print(chunk.content, flush=True)
```

## 6. 응답 메타데이터 확인

토큰 사용량 등 메타정보를 확인할 수 있습니다.

```python
response = llm.invoke("안녕")
print("content:", response.content)
print("usage:", response.usage_metadata)

#GPT 5는 reasoning 모델
```
